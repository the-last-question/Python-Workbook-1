```
#Função verificarMediana faz a comparação entre os 3 valores obtidos através dos inputs
#a fim de verificar a mediana dos valores.
def verificarMediana(a, b, c):
  if a > b:
    if a < c:
        return a
    elif b > c:
        return b
    else:
        return c
  else:
    if a > c:
        return a
    elif b < c:
        return b
    else:
        return c

a = float(input("Digite o primeiro valor: "))
b = float(input("Digite o segundo valor: "))
c = float(input("Digite o terceiro valor: "))

mediana = verificarMediana(a, b, c)

print("A mediana é: ", mediana)
```

```
#Função passwordGenerator irá gerar um número randomico para o tamanho da senha
#em seguida, serão pegos chars randomicamente ta tabela ascii entre as posições 33 e 126.

#importando a função de randomizar
import random

def passwordGenerator():
  #recebendo tamanho para a senha
  lenght = random.randint(7,10)
  password = ''

  for x in range (lenght):
    #gerando caracteres aleatórios para a senha
    randomChar = random.randint(33,126)
    password = password + chr(randomChar)

  print(password)

passwordGenerator()
```

```
#função que irá verificar a quantidade e a medida proposta para que seja mostrado a mesma unidade da forma mais otimizada o possível
def biggestMeasure(quantity, measure):
  cup = 0
  tablespoons = 0
  teaspoons = 0

#se a medida entregue for em copo apenas, deverá ser mostrado o igual o input
  if measure == "cup":
    print(str(quantity) + " Cups")

#se a medida entregue for em tablespoon devemos dividir a quantidade inicialmente por 16 para obter a quantidade de copos, em seguida por 3 para obter a quantidade de tablespoon por fim a quantidade que sobrar será a de teaspoon.
  elif measure == "tablespoon":
    if quantity >= 16:
      cup = quantity/16
      quantity = quantity % 16

    if quantity >= 3:
      tablespoons = quantity/3
      quantity = quantity % 3

    if quantity < 3:
      teaspoons = quantity
      quantity = 0

#se a medida entregue for em tablespoon devemos dividir a quantidade inicialmente por 48 para obter a quantidade de copos, em seguida por 3 para obter a quantidade de tablespoon por fim a quantidade que sobrar será a de teaspoon.
  else:
    if quantity >= 48:
      cup = quantity/48
      quantity = quantity % 48

    if quantity >= 3:
      tablespoons = quantity/3
      quantity = quantity % 3

    if quantity < 3:
      teaspoons = quantity
      quantity = 0


  print(str(int(cup)) + " Cups, " + str(int(tablespoons)) +" tablespoons " + str(int(teaspoons)) + " teaspoons.")

#criação da função main
def main():
  a = int(input("Digite a quantidade: "))
  b = input("Digite a unidade de medida: ")

  biggestMeasure(a, b)

#chamada da main
main()
```

```
#import da função de datas.
from datetime import date, timedelta
#verifica o range de datas disponíveis.
def dateRange(begin, end):
    for n in range(int ((end - begin).days)+1): yield begin + timedelta(n)
def isMagicDate(date):
  #faz o cálculo para verificar se é uma Magic Date
    if(date.day * date.month == (int(str(date.year)[2:]))): print(date.strftime("%Y-%m-%d")," is a Magic Date")
def main():
  #seta o inicio da data para o começo do século e o final para a data final.
    beginDate = date(1901,1,1)
    endDate = date(1999,12,31)
    for dates in dateRange(beginDate, endDate): isMagicDate(dates)
main()
```
